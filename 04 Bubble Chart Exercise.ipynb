```python
#######
# Objective: Create a bubble chart that compares three other features
# from the mpg.csv dataset. Fields include: 'mpg', 'cylinders', 'displacement'
# 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name'
######
```

```python
# Perform imports here:
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
# create a DataFrame from the .csv file:
df = pd.read_csv("../Data/mpg.csv")
df["horsepower"].replace({'?':50},inplace=True)
df["horsepower"]=df["horsepower"].astype(str).astype(int)
df.sort_values(by=["horsepower"],inplace=True)
df.head()
```

# Using Graph Objects

```python
#create data by choosing fields for x, y and marker size attributes

data=[go.Scatter(x=df["horsepower"], y=df["mpg"], text=df["displacement"], mode="markers", marker=dict(size=df["acceleration"], color=df["origin"], showscale=True))]
```

```python
# create a layout with a title and axis labels


layout = go.Layout(title="MPG vs Displacement",hovermode='x')



```

```python

# create a fig from data & layout, and plot the fig
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig)
```
